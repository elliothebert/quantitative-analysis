```python
# Generate batches of tensor image data with
# optional real-time data augmentation.

# preprocessing_function
# rescale=1./255 -> [0,1]
train_gen = tf.keras.layers.Rescaling(scale=1./255, offset=0.0)

train_gen = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=20, horizontal_flip=True,
width_shift_range=0.2, height_shift_range=0.2, zoom_range=0.2)

valid_gen = tf.keras.layers.Rescaling(scale=1./255, offset=0.0)
test_gen = tf.keras.layers.Rescaling(scale=1./255, offset=0.0)

train_ds = keras.utils.image_dataset_from_directory(
    directory='/content/drive/MyDrive/output/train',
    labels='inferred',
    label_mode='categorical',
    batch_size=7,
    image_size=(256, 256))
validation_ds = keras.utils.image_dataset_from_directory(
    directory='/content/drive/MyDrive/output/val',
    labels='inferred',
    label_mode='categorical',
    batch_size=2,
    image_size=(256, 256))


print(train_ds)
```

```python
def show(batch, pred_labels=None):
 plt.figure(figsize=(100, 100))
 for images, labels in batch.take(1):
  for i in range(7):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    #plt.title(class_names[labels[i]])
    plt.axis("off")


show(train_ds)
```

```python
show(validation_ds)
```

```python
model = keras.models.Sequential()
model.add(layers.Conv2D(32, (3,3), strides=(1,1), padding="valid", activation='relu', input_shape=(256, 256,3)))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(64, 3, activation='relu'))
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='softmax'))
model.add(layers.Dense(10))
print(model.summary())
```

```python
# loss and optimizer
#loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True)
#optim = keras.optimizers.Adam(lr=0.001)
#metrics = ["accuracy"]

model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
# training
#epochs = 10

# callbacks
#early_stopping = keras.callbacks.EarlyStopping(
    #monitor="val_loss",
    #patience=5,
    #verbose=2


history = model.fit(train_ds, validation_data=validation_ds, epochs=10, verbose=2)
```
